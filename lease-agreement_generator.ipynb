```python
import pandas as pd
from docx import Document
```

```python
def fill_lease_agreement(template_path, output_path, data):
    doc = Document(template_path)

    for paragraph in doc.paragraphs:
        for key, value in data.items():
            if key in paragraph.text:
                paragraph.text = paragraph.text.replace(key, str(value))

    doc.save(output_path)


def generate_lease_agreements(csv_path, template_path):
    df = pd.read_csv(csv_path, keep_default_na=False)
    for idx, row in df.iterrows():
        data = {
            "[Landlord Name]": row["landlord_name"],
            "[Mailing Address]": row["mailing_address"],
            "[Tenant Name]": row["tenant_name"],
            "[Additional Occupants]": row["additional_occupants"],
            "[Property Address]": row["property_address"],
            "[Beds]": row["beds"],
            "[Baths]": row["baths"],
            "[Start Date]": row["start_date"],
            "[End Date]": row["end_date"],
            "[Monthly Rent]": row["monthly_rent"]
        }

        output_path = '/content/drive/MyDrive/Omdena_automated_legal_compliance/data/contract_'+str(idx)+'.docx'

        fill_lease_agreement(template_path, output_path, data)
```

```python
csv_path = '/content/drive/MyDrive/Omdena_automated_legal_compliance/lease_agreements_50_rows.csv'
template_path = '/content/drive/MyDrive/Omdena_automated_legal_compliance/residential_lease_agreement.docx'
generate_lease_agreements(csv_path, template_path)
```
